```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('iris')
df.head()
```

```python
df.isnull().sum()
```

```python
sns.scatterplot(x='sepal_length',y='sepal_width', data=df, hue='species')
plt.show()
```

```python
sns.scatterplot(x='petal_length',y='petal_width', data=df, hue='species')
plt.show()
```

```python
sns.pairplot(df, hue='species', height=2.0)
plt.show()
```

```python
columns = ['sepal_length','sepal_width','petal_length','species']
sns.pairplot(df[columns], hue='species', height=2.0)
plt.show()
```

```python
X = df.drop('species', axis=1)
```

```python
feature_cols = ['sepal_length','sepal_width','petal_length','petal_width']
X = df[feature_cols]
X.head()
```

```python
Y = df.species
Y.head()
```

```python
X.shape
```

```python
Y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X, Y, test_size=0.3)
```

```python
import numpy as np
(label, count) = np.uinque(Y_test , return_counts=True)
freq = np.asarray((label, count)).T
print(freq)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
from sklearn.svm import SVC

model = SVC(kernel= 'linear')

model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print('Accuracy Score: [:.4f]' .format(accuracy_score(Y_test, y_pred)))
print(classification_report(Y_test, y_pred ))
cm = confusion_matrix(Y_test, y_pred)
print(cm)
```
